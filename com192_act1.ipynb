### 2. Funciones

- Una función es un grupo de instrucciones autocontenidas dentro del programa que realiza una tarea única.
- Las funciones ayudan al programador a resolver un problema de manera más simple.
- En lugar de escribir un programa que resuelve un problema grande, se escriben varias funciones que resuelven problemas pequeños.

#### Ejemplos

###### 1. Calcular el área de un rectángulo

```python
def area_rectangulo():

    # pido variables
    altura = int(input("dame la altura: "))
    base = int(input("dame la base: "))

    # calcular área del rectángulo
    area = base * altura

    return area
    #print ("res: ", area)
    #print("\n")

res = area_rectangulo()
print ("res: ", res)
#print("\n")
#area_rectangulo()
```

##### 2. Calcular el área de un triángulo

```python
def main():
  a, b = pido_variables() # a, b
  res = area_triangulo(b, a)
  print ("res: ", res)
  print("\n")

def pido_variables():

    # pido variables
    a = int(input("dame a: "))
    b = int(input("dame b: "))

    return a, b

def area_triangulo(base = 1, altura = 1):

    # calcular área del triángulo
    return (base * altura)/2

main()
```

##### 3. Costo del terreno

```python
def dame_variables():

    # pido variables
    a = int(input("dame a: "))
    b = int(input("dame b: "))
    c = int(input("dame c: "))
    d = int(input("dame d: "))

    return a, b, c, d

def area_rect(base = 1, altura = 1):

    # calcular área del rectángulo
    area = base * altura

    return area

def area_trian(base = 1, altura = 1):

    # calcular área del triángulo
    area = (base * altura)/2

    return area

def costo_total(area1, area2, area3):

    # calcular área total del terreno
    att = area1 - area2 - area3
    # calcular valor del terreno
    vtt = att * 3450

    return vtt

def codigo_costo_terreno():

    # pido variables
    a, b, c, d = dame_variables()
    # calculo las áreas
    a1 = area_rect(a, b)
    a2 = area_trian(b, d)
    a3 = area_trian(b, c)
    # calculo el costo
    vtt = costo_total(a1, a2, a3)

    print("el valor del terreno es: $", vtt)

codigo_costo_terreno()
```

```python
# un programa que pide 2 números y encuentra el mayor
# un ejemplo de cómo implemento if

def buscaMayor():
    x = int(input("dame x: ")) # el dato lo convierto de texto a número
    y = int(input("dame y: "))

    if y < x:
        print("x es mayor")
    elif y == x:
        print("son iguales")
    else:
        print("y es mayor")

#def main():

#    print("Hola")
#    buscaMayor()

buscaMayor()
```

### 5. Graficar

```python
import matplotlib.pyplot as plt # to plot
```

```python
def generaDatos():

    df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))

    return df

def graficaDatos(datos):

    plt.figure()
    plt.plot(datos["A"],label="Datos A", color='red')
    plt.grid(color='b',linestyle='dashed')
    plt.title('DATOS', fontweight='bold')
    plt.xlabel('Tiempo')
    plt.ylabel('Valor A')
    plt.legend(loc='upper left')
    plt.show()

df = generaDatos()
graficaDatos(df)
```

```python
def graficaValores(datos):

    plt.figure()
    plt.plot(datos["B"].sort_values(), datos["A"], label="Datos A", color='orange')
    plt.grid(color='b',linestyle='dashed')
    plt.title('DATOS', fontweight='bold')
    plt.xlabel('Valor B')
    plt.ylabel('Valor A')
    plt.legend(loc='upper right')
    plt.show()

graficaValores(df)
```
